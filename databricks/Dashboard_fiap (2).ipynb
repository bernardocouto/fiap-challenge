```
%sql
select date_format(date ,'y') Ano, sum (new_confirmed) Casos  from default.covid19 
where country_code='US' AND subregion1_code='IA' and new_confirmed 
is not null group by date_format(date ,'y') order by 2 desc ;
```

```
%sql
select Date_format(TO_DATE(CAST(UNIX_TIMESTAMP(date, 'MM/dd/yyyy') AS TIMESTAMP)),'y') Ano, round(sum(Volume_Sold_Liters),0) Volume_Liters from default.iowa_liquor 
where Date_format(TO_DATE(CAST(UNIX_TIMESTAMP(date, 'MM/dd/yyyy') AS TIMESTAMP)),'y') between '2017' and '2022'
group by Date_format(TO_DATE(CAST(UNIX_TIMESTAMP(date, 'MM/dd/yyyy') AS TIMESTAMP)),'y')
order by 2 desc
limit 10
```
